#Blue Chip Portfolio

```python
tickers = ["AAPL", "MSFT", "JNJ", "JPM", "PG"]
data = yf.download(tickers, start="2005-01-01", end="2026-03-05", interval="1d")
df = data["Close"]
df = df.pct_change(1)
df = df.dropna()
df = df.reset_index(drop=True)
df.to_csv("/content/drive/MyDrive/Masterthesis/Data/bluechip_sp500.csv", index=False)
```

```python
tickers = ['SPY', 'VUSTX', 'VGSIX', 'GC=F', 'CL=F']
data = yf.download(tickers, start="2000-08-30", end="2024-09-27", interval="1d")
df = data["Close"].copy()
rename_dict = {
    'SPY': 'SP500',
    'VUSTX': 'Long_Treasuries',
    'VGSIX': 'Real_Estate',
    'GC=F': 'Gold',
    'CL=F': 'Crude_Oil'
}
df = df.rename(columns=rename_dict)
df = df[['SP500', 'Long_Treasuries', 'Real_Estate', 'Gold', 'Crude_Oil']]
df = df.pct_change(1)
df = df.dropna()
df = df.reset_index(drop=True)
df.to_csv("/content/drive/MyDrive/Masterthesis/Data/crossAsset_portfolio.csv", index=False)
```
